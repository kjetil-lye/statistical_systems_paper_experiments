# Fractional Brownian motion plots

In this notebook, we plot the fractional Brownian motion plots.

```python
# %load defaults.ipy

import numpy as np
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 600
%matplotlib inline

import matplotlib.pyplot as plt

import sys
import os
import os.path
sys.path.append('../python')
import plot_info
from plot_info import showAndSave
import netCDF4
from IPython.core.display import display, HTML
from mpl_toolkits.mplot3d import Axes3D
import copy
import json
import traceback
plot_info.set_notebook_name("FractionalBrownianMotion.ipynb")

sys.path.append('fbm')
import euler_brownian

import random
def seed_random_number(seed):
    # see https://stackoverflow.com/a/52897216
    np.random.seed(seed)
   
    random.seed(seed)
    
seed_random_number(0)
```

```python
N = 1024
H = 0.5

x, y = np.mgrid[0:1:N*1j, 0:1:N*1j]

# We have two fields:
for direction in ['x', 'y']:
    X = np.random.normal(0,1,N**2)
    
    u = euler_brownian.fBm(N, H, euler_brownian.RandomVariable(X))[:-1,:-1]
    
    fig = plt.figure(figsize=(10,8))
    ax = fig.gca(projection='3d')
    
    ax.plot_surface(x, y, u)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel("$w^{{{},0.5}}_{{0}}$".format(direction))
    showAndSave('brownian2d_u{}'.format(direction))
    
```
